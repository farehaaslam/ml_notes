```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
X.shape
```

```python
def gd(X,y):
    X=np.insert(X,0,1,axis=1)   #coff of x0 ie consatnt term
    w=np.ones(X.shape[1])   #initail weight matrice 
    lr=0.5
    for i in range(10000):
        y_cap=sigmoid(np.dot(X,w))
        w=w+lr*np.dot((y-y_cap),X)/X.shape[0]

    return w[1:,],w[0]  
```

```python
coeff,intercept=gd(X,y)
```

```python
coeff
```

```python
intercept
```

```python
m = -(coeff[0]/coeff[1])
b = -(intercept/coeff[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b 
```
