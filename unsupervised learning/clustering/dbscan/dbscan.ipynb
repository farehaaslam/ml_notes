```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('Mall_Customers.csv')
```

```python
X=np.array([[2,2],[2,2.5],[7,8],[4,4],[4,6],[4,5]])
```

```python
plt.scatter(X[:,0],X[:,1])
plt.plot()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1,min_samples=2)
dbscan.fit_predict(X)
```
